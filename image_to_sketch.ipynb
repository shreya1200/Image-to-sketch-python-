```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('minion.jpg')
# img = cv2.imread('index.jpeg', cv2.IMREAD_GRAYSCALE)
original_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(original_img)
```

```python
plt.axis("off")
cartoon_image = cv2.stylization(original_img, sigma_s=150, sigma_r=0.25)  
plt.imshow(cartoon_image) 
cv2.imwrite('sketch1.png',cartoon_image)
```

```python
plt.axis("off")
gray_img = cv2.cvtColor(original_img.astype(np.uint8),cv2.COLOR_BGR2GRAY)
# cv2.imshow("Converted Image",gray_img)
plt.imshow(gray_img, cmap="gray", vmin=0, vmax=255)
```

```python
plt.axis("off")
img_invert = cv2.bitwise_not(gray_img)
plt.imshow(img_invert, cmap="gray", vmin=0, vmax=255)
```

```python
plt.axis("off")
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.imshow(img_smoothing, cmap="gray", vmin=0, vmax=255)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=250)
```

```python
plt.axis("off")
final_img = dodgeV2(gray_img, img_smoothing)
plt.imshow(final_img, cmap="gray", vmin=240, vmax=255)
cv2.imwrite('sketch.png',final_img)
```
